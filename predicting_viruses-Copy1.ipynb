### Predict species based on other variables:

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
#Assumed you have, X (predictor) and Y (target) for training data set and x_test(predictor) of test_dataset
knowndf = df.query("species != 'unclass'")
unknowdf= df.query(("species == 'unclass'"))
# Create Random Forest object
```

```python
model= RandomForestClassifier(n_estimators=1000)
```

```python
knowndf.head()
```

```python
from sklearn.cross_validation import train_test_split
train_data,test_data = train_test_split(knowndf,train_size=0.8)
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
y, _ = pd.factorize(train_data['species'])
```

```python
model.fit(train_data[features], y)
```

```python
preds = model.predict(test_data[features])
#y, _ = pd.factorize(train['species'])
```

```python
pd.crosstab(test_data['species'], preds, rownames=['actual'], colnames=['preds'])
```
